```python
import pandas as pd
```

```python
df = pd.read_csv('./netflix_titles.csv')
```

```python
df.isnull().sum()
```

```python
# Create a mapping of directors to casts
director_cast_mapping = df[df['director'].notna()].set_index('cast')['director'].to_dict()

# Populate null values in the 'director' column based on the 'cast' column
df['director'] = df['director'].fillna(df['cast'].map(director_cast_mapping))

# Fill remaining null values in the 'director' column with 'Not Given'
df['director'] = df['director'].fillna('Not Given')

# Check null values in 'director' column
df['director'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Create a mapping of countries to directors
country_director_mapping = df[df['country'].notna()].set_index('director')['country'].to_dict()
country_director_mapping
```

```python
# Populate null values in the 'country' column based on the 'director' column
df['country'] = df['country'].fillna(df['director'].map(country_director_mapping))

# Fill remaining null values in the 'country' column with 'Not Given'
df['country'] = df['country'].fillna('Not Given')

# Check null values in 'country' column
df['country'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
cast_director_mapping = df[df['cast'].notna()].set_index('director')['cast'].to_dict()
cast_director_mapping
```

```python
df['cast'] = df['cast'].fillna(df['director'].map(cast_director_mapping))
df['cast'] = df['cast'].fillna('Not Given')
```
