<h1 style="font-size:2em;color:#2467C0">Data Engineering: Step 2A, Exploring Data</h1>

## Data Ingestion, Cleaning
In this notebook, we will use three CSV files: ratings.csv, tags.csv, and movies.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

movies_df = pd.read_csv('./movielens/movies.csv', sep=',')
tags_df = pd.read_csv('./movielens/tags.csv', sep=',')
ratings_df = pd.read_csv('./movielens/ratings.csv', sep=',')

movies_df.head(), tags_df.head(), ratings_df.head()
```

```python
# Drop unnecessary columns
movies_df.drop('title', axis=1, inplace=True)
tags_df.drop(['userId', 'timestamp'], axis=1, inplace=True)
ratings_df.drop(['userId', 'timestamp'], axis=1, inplace=True)

movies_df.head(), tags_df.head(), ratings_df.head()
```

<h1 style="font-size:2em;color:#2467C0">Data Engineering: Step 2B, Pre-Processing Data</h1>

```python
# Check for missing values
movies_df.isnull().any(), tags_df.isnull().any(), ratings_df.isnull().any()
```

```python
# Remove any missing values from the tags_df
tags_df.dropna(inplace=True)

tags_df.isnull().any()
```

## Data Merging

```python
# Calculate the average rating for each movie and overall
average_ratings_df = ratings_df.groupby('movieId')['rating'].mean().reset_index()
average_rating = ratings_df['rating'].mean()

average_ratings_df, average_rating
```

```python
# Merge tags_df and movies_df
merged_df = pd.merge(tags_df, movies_df, on='movieId', how='inner')

# Combine the tags for each movie
merged_df = merged_df.groupby(['movieId', 'genres'])['tag'].agg('|'.join).reset_index()

merged_df
```

```python
# Merge with the average ratings dataframe and remove duplicates
merged_df = pd.merge(merged_df, average_ratings_df, on='movieId', how='inner')
merged_df.drop_duplicates(subset='movieId', inplace=True)

merged_df
```

## Data Filtering

```python
# Filter out movies with no listed genres and movies with a rating less than the average rating
merged_df = merged_df[(merged_df['genres'] != '(no genres listed)') & (merged_df['rating'] > average_rating)]

# Reset index
merged_df.reset_index(drop=True, inplace=True)

merged_df
```

## Vectorized String Operations

```python
# Split 'genres' and 'tag' into lists
merged_df = merged_df.copy()
merged_df['genres'] = merged_df['genres'].str.split('|')
merged_df['tag'] = merged_df['tag'].str.split('|')

merged_df
```

<h1 style="font-size:2em;color:#2467C0">Data Analysis: Step 3, Analyze Data</h1>

## Data Transformation and Feature Engineering

```python
from itertools import product

# Generate all possible combinations of genres and tags for each movie
merged_df['genre_tag'] = merged_df.apply(lambda row: list(product(row['genres'], row['tag'])), axis=1)

merged_df
```

```python
# Explode the dataframe on the 'genre_tag' column
exploded_df = merged_df.explode('genre_tag')

exploded_df
```

```python
# Split the 'genre_tag' column into two separate columns
exploded_df[['genre', 'tag']] = pd.DataFrame(exploded_df['genre_tag'].tolist(), index=exploded_df.index)

# Drop the 'genre_tag' column
exploded_df = exploded_df.drop('genre_tag', axis=1)

exploded_df
```

```python
# Calculate the total number of ratings and the average rating for each genre-tag combination
genre_tag_df = exploded_df.groupby(['genre', 'tag']).agg({'rating': ['count', 'mean']}).reset_index()

# Flatten the column names
genre_tag_df.columns = ['genre', 'tag', 'rating_count', 'rating_mean']

genre_tag_df
```

```python
# Calculate the overall average rating count and mean rating
average_rating_count = genre_tag_df['rating_count'].mean()
average_rating_mean = genre_tag_df['rating_mean'].mean()

average_rating_mean
```

```python
# Identify genre-tag combinations with a higher total number of ratings and a higher average rating than the overall average
popular_high_rated_combinations = genre_tag_df[(genre_tag_df['rating_count'] > average_rating_count) & (genre_tag_df['rating_mean'] > average_rating_mean)]

popular_high_rated_combinations
```

```python
# Sort the dataframe by rating count in descending order
popular_high_rated_combinations = popular_high_rated_combinations.copy()
popular_high_rated_combinations.sort_values('rating_count', ascending=False, inplace=True)

# Reset the index of the dataframe
popular_high_rated_combinations.reset_index(drop=True, inplace=True)

popular_high_rated_combinations
```

```python
# Display the top 10 genre-tag combinations for visualization
top_10_combinations = popular_high_rated_combinations.head(10)

top_10_combinations
```

```python
# Sort the top 10 combinations by average rating in descending order
top_10_combinations_sorted = top_10_combinations.sort_values('rating_mean', ascending=False)

top_10_combinations_sorted
```

<h1 style="font-size:2em;color:#2467C0">Data Analysis: Step 4, Reporting Insights</h1>

```python
# Create the bar plot for rating counts
plt.figure(figsize=(14, 7))
plt.barh(top_10_combinations['genre'] + '-' + top_10_combinations['tag'], top_10_combinations['rating_count'], color='skyblue')
plt.xlabel('Total Number of Ratings')
plt.ylabel('Genre-Tag Combination')
plt.title('Top 10 Genre-Tag Combinations by Total Number of Ratings')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Create the bar plot for average ratings
plt.figure(figsize=(14, 7))
plt.barh(top_10_combinations_sorted['genre'] + '-' + top_10_combinations_sorted['tag'], top_10_combinations_sorted['rating_mean'], color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Genre-Tag Combination')
plt.title('Top 10 Genre-Tag Combinations by Average Rating')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Merge top_10_combinations and top_10_combinations_sorted dataframes on genre and tag
combined_df = pd.merge(top_10_combinations, top_10_combinations_sorted, on=['genre', 'tag'], suffixes=('_count', '_mean'))
# Flip the dataframe
combined_df_flipped = combined_df[::-1]

fig, ax1 = plt.subplots(figsize=(14, 7))
ax1.barh(combined_df_flipped['genre'] + '-' + combined_df_flipped['tag'], combined_df_flipped['rating_count_count'], color='skyblue')
ax1.set_xlabel('Total Number of Ratings')
ax1.set_ylabel('Genre-Tag Combination')
# Create a second y-axis to plot average ratings
ax2 = ax1.twiny()
ax2.plot(combined_df_flipped['rating_mean_mean'], combined_df_flipped['genre'] + '-' + combined_df_flipped['tag'], color='darkorange', marker='o')
ax2.set_xlabel('Average Rating')
plt.title('Top 10 Genre-Tag Combinations by Total Number of Ratings and Average Rating')
plt.show()
```
